# Ejercicio 1: Complejidad Lineal O(n) - Búsqueda Lineal

```python
def busqueda_lineal(array, objetivo):
    for i in range(len(array)):
        if array[i] == objetivo:
            return i
    return -1

arr = [7, 9, 10, 12, 1, 0]
valor = 9

resultado = busqueda_lineal(arr, valor)
print("Resultado:", resultado)
```

# Mejor caso

Ocurre cuando el elemento buscado se encuentra en la primera posición.

- Comparaciones: 1
- Retorno: 1

$$
T_{mejor}(n) \in \Theta(1)
$$


#Peor caso
Ocurre cuando el elemento buscando se encuentra en la posición n del array o no existe.

- Comparaciones n + 1
- Retorno: 1

$$
T_{peor}(n) = n + 1
$$

Eliminando constantes:

$$
T_{peor}(n) \in \Theta(n)
$$

# Caso promedio
Si el elemento buscado se encuentra en la posición k, se realizan k + 1 comparaciones.

El elemento puede estar en cualquiera de las n posiciones del array, entonces:
$$
T_{prom}(n) = \frac{1}{n}\sum_{k=0}^{n-1}(k+1)
$$

Sabemos que:
$$1 + 2 + 3 + \dots + n = \frac{n(n+1)}{2}
$$

Entonces:

$$
T_{prom}(n) = \frac{1}{n} \cdot \frac{n(n+1)}{2}
$$


Simplificando:

$$
T_{prom}(n) = \frac{n+1}{2}
$$

Por lo tanto:

$$
T_{prom}(n) \in \Theta(n)
$$

# Orden de complejidad al que pertenece
La búsqueda lineal pertenece al orden de complejidad lineal:

$$
O(n)
$$

# Ejercicio 2: Complejidad Logarítmica O(log n) — Búsqueda Binaria


```python
def busqueda_binaria(arr, objetivo):
    izquierda = 0
    derecha = len(arr) - 1

    while izquierda <= derecha:
        medio = (izquierda + derecha) // 2

        if arr[medio] == objetivo:
            return medio
        elif arr[medio] < objetivo:
            izquierda = medio + 1
        else:
            derecha = medio - 1

    return -1
```

## Mejor caso
Ocurre cuando el elemento buscado se encuentra en la mitad del array, en la primera iteración.

$$
T_{mejor}(n) \in \Theta(1)
$$

## Peor caso
Ocurre cuando el elemento no está o se encuentra en el último nivel de las divisiones del array.

En cada iteración el tamaño del problema se reduce a la mitad:

$$
\frac{n}{2^k}
$$

El algoritmo termina cuando el tamaño se reduce a 1:

$$
\frac{n}{2^k} = 1
$$

Entonces:

$$n = 2^k$$

Aplicando logaritmo base 2 de ambos lados:

$$k = \log_2(n)$$

Ahora para el peor caso:

$$
T_{peor}(n) = a\log_2(n) + b
$$
\
($a$ son las operaciones dentro del while, que se realizarán $log_2(n)$ veces, $b$ son las operaciones que se realizan antes y después del while, se realizan 1 sola vez cada una)


Eliminando constantes:

$$
T_{peor}(n) \in \Theta(\log n)
$$

## Caso promedio

En promedio también se requieren aproximadamente $ \log(n) $ iteraciones.

$$
T_{prom}(n) \in \Theta(\log n)
$$

##Orden de complejidad al que pertenece

La búsqueda binaria pertenece al orden de complejidad:

$$
O(\log n)
$$

# Ejercicio 3: Complejidad Cuadrática O(n²) - Ordenamiento Burbuja

```python
def burbuja(arr):
    n = len(arr)
    for i in range(n):
        swapped = False
        for j in range(0, n - i - 1):
            if arr[j] > arr[j + 1]:
                arr[j], arr[j + 1] = arr[j + 1], arr[j]
                swapped = True
        if not swapped:
            break
    return arr
```

## Mejor caso
Si el array ya está ordenado, solo se recorre una vez.
\
La operación principal es $ arr[j] > arr[j + 1].$

$$
T_{mejor}(n) = n-1
$$

$$
T_{mejor}(n) \in \Theta(n)
$$

##Peor caso
Posiciones del array en orden inverso (descendente).
El número total de comparaciones está dado por:

$$
T(n) = \sum_{i=0}^{n-1} (n - i - 1)
$$
Se recorre el array la mayor cantidad de veces posible, entonces:
$$
T_{peor}(n) = \frac{n(n-1)}{2}
$$

$$
T_{peor}(n) \in \Theta(n^2)
$$

## Caso promedio

El número de comparaciones no depende del orden del array, por lo tanto:

$$
T_{prom}(n) = \frac{n(n-1)}{2}
$$

$$
T_{prom}(n) \in \Theta(n^2)
$$

## Orden de complejidad al que pertenece
El ordenamiento burbuja pertenece al orden de complejidad:

$$
O(n^2)
$$

## Ejercicio 4: Complejidad Cúbica O(n³) - Multiplicación de Matrices

```python
def mult_matrices(A, B):
    n = len(A)
    C = [[0 for _ in range(n)] for _ in range(n)]

    for i in range(n):
        for j in range(n):
            suma = 0
            for k in range(n):
                suma += A[i][k] * B[k][j]
            C[i][j] = suma

    return C
```

La operación dominante se realiza en el bucle interno:

$$
suma \mathrel{+}= A[i][k]\cdot B[k][j]
$$

Conteo de iteraciones:
- El bucle \(i\) se ejecuta \(n\) veces.
- El bucle \(j\) se ejecuta \(n\) veces por cada \(i\).
- El bucle \(k\) se ejecuta \(n\) veces por cada par \((i,j)\).
\


$$
n \cdot n \cdot n = n^3
$$

Si las operaciones dentro de cada operación cuestan \(c\):

$$
T(n) = c\,n^3 + b
$$

Por lo tanto:

$$
T(n) \in \Theta(n^3)
$$

## Casos

$$
n \cdot n \cdot n = n^3
$$

Si las operaciones dentro de cada iteración cuestan \(c\):

$$
T(n) = c\,n^3 + b
$$

Por lo tanto:
$$
T(n) \in \Theta(n^3)
$$

## Orden del complejidad al que pertenece
La multiplicación de matrices pertenece al orden de complejidad:
$$\Theta(n^3)$$

El ciclo se ejecuta 100 veces:

C_1 = 100

Dos bucles anidados:

- El bucle interno se ejecuta \(n\) veces por cada iteración del externo.

Total:

$$
C_2 = n \cdot n = n^2
$$

---
## Segmento 3

El ciclo se ejecuta \(n\) veces:

$$
C_3 = n
$$


# Expresión Completa

Sumamos los tres segmentos:

$$
T(n) = 100 + n^2 + n
$$

Aplicando la regla de la suma, el término dominante es:

$$
n^2
$$
Por lo tanto:

$$
T(n) \in \Theta(n^2)
$$

## Orden de complejidad final:

$$
O(n^2)

#Funcion medir_tiempo:

```python
#Implementación:
import time
import matplotlib.pyplot as plt
import random
def medir_tiempo(algoritmos, *args):
  inicio = time.time()
  contador = algoritmos(*args)
  fin = time.time()
  return fin - inicio, contador
```

# Experimento busqueda lineal O(n):

```python
tamaños = [10, 50, 100, 500, 1000, 5000]

tiempos_experimentales = []
contadores = []

for n in tamaños:
    arr = [random.randint(1, 10000) for _ in range(n)]
    objetivo = arr[-1]  # peor caso

    tiempos = []

    for _ in range(5):  # repetir 5 veces
        tiempo, contador = medir_tiempo(busqueda_lineal, arr, objetivo)
        tiempos.append(tiempo)

    tiempo_promedio = sum(tiempos) / 5

    tiempos_experimentales.append(tiempo_promedio)
    contadores.append(contador)

    print(f"n = {n}")
    print(f"Tiempo promedio = {tiempo_promedio}")
    print(f"Contador = {contador}")
    print("-----------")
```
